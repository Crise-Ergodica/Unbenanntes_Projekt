### Configurando o Setup de Regressão

```python
reg1 = setup(data=dfTrain,
             target='SalePrice',
             session_id=42, # Usando 42 para mudar a semente

             ignore_features=['Id'], # IGNORAR O ID DO IMÓVEL

             transform_target=True, # TRANSFORMAÇÃO DO TARGET (y, SalePrice) PARA DISTRIBUIÇÃO NORMAL


             imputation_type='simple', # TRATAMENTO DE VALORES AUSENTES

             # OUTLIERS
             # Remoção de Outliers (opcional, mas bom para regressão)
             # remove_outliers=True,

             # VALIDAÇÃO CRUZADA
             # Usando 5 folds para agilizar a comparação inicial
             fold=5
            )
```

```python
ignore_list = ['lightgbm'] #
best_model = compare_models(turbo = True, exclude= ignore_list)
# turbo é usado para acelerar o processo, mas pode sacrificar a precisão. Por isso não devo usar em produção.
```

### Criando o Modelo GBR

```python
modeloGBR = create_model('gbr') # Criando o Modelo

tuned_gbr = tune_model(modeloGBR,
                       optimize = 'RMSE',
                       n_iter = 20) # Optimizando/Tunning o Modelo
```

```python
final_gbr = finalize_model(tuned_gbr)
save_model(final_gbr, 'housePricesGBR')
```
