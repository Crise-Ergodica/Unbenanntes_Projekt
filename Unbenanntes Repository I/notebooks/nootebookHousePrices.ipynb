### Estruturando o básico
Coisas que são necessárias e/ou vamos usar até o final do código. Algumas bibliotecas e o CSV.

```python
import pandas as pd
from pycaret.regression import *
dfTrain = pd.read_csv('../data/raw/house-prices-advanced-regression-techniques/train.csv')
dfTest = pd.read_csv('../data/raw/house-prices-advanced-regression-techniques/test.csv')
```

_Id não tem valor preditivo, então vamos ignorar, além disso SalePrice é o Target (preço da casa)._


### Configurando o Setup de Regressão

```python
reg1 = setup(data=dfTrain,
             target='SalePrice',
             session_id=42, # Usando 42 para mudar a semente

             ignore_features=['Id'], # IGNORAR O ID DO IMÓVEL

             transform_target=True, # TRANSFORMAÇÃO DO TARGET (y, SalePrice) PARA DISTRIBUIÇÃO NORMAL


             imputation_type='simple', # TRATAMENTO DE VALORES AUSENTES

             # OUTLIERS
             # Remoção de Outliers (opcional, mas bom para regressão)
             remove_outliers=True,

             # VALIDAÇÃO CRUZADA
             # Usando 5 folds para agilizar a comparação inicial
             fold=5
            )
```
